Displacement VS residue

```python
import pandas as pd

# Read the file, skipping the comment lines at the top
df = pd.read_csv(f'/user/work/fh22600/1A5_Sim/vec_norm_stats/550_1A5_298_strip_6NB_vec_norm_10.txt', skiprows=5, delim_whitespace=True, header=None)


# CA first column against average displacement 8th column
#12 and 16 are 1Se and 2SE

## 10 repeats
## top plot graph of ca atoms vs average displacement
a_disp = df.iloc[:,7]
ca =df.iloc[:,0]
yerr = df.iloc[:,15]

plt.figure(figsize=(20, 6))
#plt.plot(ca, a_disp,color='blue', label='Data', marker='o')



mutations = [23, 148, 182, 219, 237, 238, 248, 253, 257]
mutations_y = np.interp(mutations, ca, a_disp)
plt.plot(ca, a_disp, linewidth=1.5)
plt.plot(mutations, mutations_y, 'bx', label='Mutation')

plt.legend(loc='center', fontsize='small')

plt.axhline(y=0, color='green', linestyle='--', linewidth=1, label='y=0')
plt.errorbar(ca,a_disp, yerr=yerr, fmt='none', ecolor='red',capsize=5, label='Displacements with SD')
plt.title('2SD Avg. Disp. VS residue for 10 repeats')
plt.xlabel('Residue')
plt.ylabel('Displacements')
```

```python
##plot of 5rep and 10rep displacements

import pandas as pd

# Read the file, skipping the comment lines at the top
allrep = pd.read_csv(f'vec_norm_stats/180_1A5_298_strip_6NB_vec_norm_10.txt', skiprows=5, delim_whitespace=True, header=None)
fiverep = pd.read_csv(f'vec_norm_stats/370_1A5_298_strip_6NB_vec_norm_10.txt', skiprows=5, delim_whitespace=True, header=None)
wholedata = pd.read_csv(f'vec_norm_stats/550_1A5_298_strip_6NB_vec_norm_10.txt', skiprows=5, delim_whitespace=True, header=None)
# CA first column against average displacement 8th column
#12 and 16 are 1Se and 2SE

## 10 repeats
## top plot graph of ca atoms vs average displacement
all_disp = allrep.iloc[:,7]
five_disp = fiverep.iloc[:,7]
whole_disp = wholedata.iloc[:,7]
ca = allrep.iloc[:,0] 
allyerr = allrep.iloc[:,11]
fiveyerr = fiverep.iloc[:,11] ## 2SD of average displacement
wholeyerr = wholedata.iloc[:,11]

plt.figure(figsize=(20, 6))
#plt.plot(ca, a_disp,color='blue', label='Data', marker='o')
plt.errorbar(ca,all_disp, yerr=allyerr, fmt='-o', ecolor='blue',capsize=5, label='180')
plt.errorbar(ca,five_disp, yerr=fiveyerr, fmt='-s', ecolor='orange',capsize=5, label='370 ')
plt.errorbar(ca,whole_disp, yerr=wholeyerr, fmt='-x', ecolor='red',capsize=5, label='550')

mutations = [23, 148, 182, 219, 237, 238, 248, 253, 257]
mutations_y = np.interp(mutations, ca, all_disp)
#plt.plot(ca, a_disp, linewidth=1.5)
#plt.plot(mutations, mutations_y, 'bs', label='Mutation')

plt.legend(fontsize='small')

plt.axhline(y=0, color='green', linestyle='--', linewidth=1, label='y=0')
plt.ylim(-0.2,0.8)
plt.title('1A53-2.5: Avg Displacement (1SD) with increased sampling')
plt.xlabel('Residue')
plt.ylabel('Cα displacements / Å')
plt.ylim(-0.3,0.8)
```
