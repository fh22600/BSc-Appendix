Comparing at different times

```python
timepoints = [10,20,30,40]
fig, axes = plt.subplots(1, 4, figsize=(20, 5))
axes = axes.flatten()
for i, tp in enumerate(timepoints):
    null = pd.read_csv(f'/user/work/fh22600/1A0_Sim/vec_norm_txt/550_1SD_1A0_298_null_{tp}.txt', skiprows=5, delim_whitespace=True, header=None)
    perturbed = pd.read_csv(f'//user/work/fh22600/1A0_Sim/vec_norm_txt/550_1SD_1A0_298_strip_6NB_vec_norm_{tp}.txt', skiprows=5, delim_whitespace=True, header=None)
    # CA first column against average displacement 8th column
    #12 and 16 are 1Se and 2SE
    # 1SD
    ## 10 repeats
    ## top plot graph of ca atoms vs average displacement
    null_disp = null.iloc[:,7]
    perturbed_disp = perturbed.iloc[:,7]
    ca = null.iloc[:,0] 
    #nullerr = null.iloc[:,11]
    #perturbederr = perturbed.iloc[:,11] ## 1SD of average displacement

    ax = axes[i]
    ax.errorbar(ca, null_disp, color='green', ecolor='green',capsize=5, linewidth=1.5, label=f'{tp} null')
    ax.errorbar(ca, perturbed_disp, color='red', ecolor='red',capsize=5, linewidth=1.5, label=f'{tp} perturbed')
    ax.set_xlabel('Residue')
    ax.set_ylabel('Average displacements / ')
    ax.set_title(f'1A53-2: Avg. Disp and null p. after {tp} ps')
    ax.legend(loc='upper center', fontsize='small')
plt.tight_layout()
```
