Checking displacements with 2SD.

```python
#  all data and error bars

import pandas as pd

# Read the file, skipping the comment lines at the top
df = pd.read_csv(f'/user/work/fh22600/1A0_Sim/vec_norm_txt/550_1A0_298_strip_6NB_vec_norm_10.txt', skiprows=5, delim_whitespace=True, header=None)


## 10 repeats
## top plot graph of ca atoms vs average displacement
a_disp = df.iloc[:,7]
ca =df.iloc[:,0]
yerr = df.iloc[:,15]

plt.figure(figsize=(20, 6))
#plt.plot(ca, a_disp,color='blue', label='Data', marker='o')



mutations = [23, 148, 182, 219, 237, 238, 248, 253, 257]
mutations_y = np.interp(mutations, ca, a_disp)
plt.plot(ca, a_disp, linewidth=1.5)
plt.plot(mutations, mutations_y, 'bx', label='Mutation')

plt.legend(loc='center', fontsize='small')

plt.axhline(y=0, color='green', linestyle='--', linewidth=1, label='y=0')
plt.errorbar(ca,a_disp, yerr=yerr, fmt='none', ecolor='red',capsize=5, label='Displacements with SD')
plt.title('2SD Avg. Disp. VS residue for 10 repeats')
plt.xlabel('Residue')
plt.ylabel('Displacements')
```

Error analysis with 2SD

```python
##plot of 5rep and 10rep displacements

import pandas as pd

# Read the file, skipping the comment lines at the top
firstrep = pd.read_csv(f'/user/work/fh22600/1A0_Sim/vec_norm_txt/180_1A0_298_strip_6NB_vec_norm_100.txt', skiprows=5, delim_whitespace=True, header=None)
secondrep = pd.read_csv(f'/user/work/fh22600/1A0_Sim/vec_norm_txt/370_1A0_298_strip_6NB_vec_norm_100.txt', skiprows=5, delim_whitespace=True, header=None)
thirdrep = pd.read_csv(f'/user/work/fh22600/1A0_Sim/vec_norm_txt/550_1A0_298_strip_6NB_vec_norm_100.txt', skiprows=5, delim_whitespace=True, header=None)

#12 and 16 are 1Se and 2SE

## 10 repeats
## top plot graph of ca atoms vs average displacement
first_disp = firstrep.iloc[:,7]
second_disp = secondrep.iloc[:,7]
third_disp = thirdrep.iloc[:,7]
ca = firstrep.iloc[:,0] 
firstyerr = firstrep.iloc[:,15]
secondyerr = secondrep.iloc[:,15] ## 2SD of average displacement
thirdyerr = thirdrep.iloc[:,15]

plt.figure(figsize=(20, 6))
#plt.plot(ca, a_disp,color='blue', label='Data', marker='o')
plt.errorbar(ca,first_disp, yerr=firstyerr, fmt='-o', ecolor='blue',capsize=5, label='180')
plt.errorbar(ca,second_disp, yerr=secondyerr, fmt='-s', ecolor='orange',capsize=5, label='370')
plt.errorbar(ca,third_disp, yerr=thirdyerr, fmt='-x', ecolor='red',capsize=5, label='550')

#mutations = [23, 148, 182, 219, 237, 238, 248, 253, 257]
#mutations_y = np.interp(mutations, ca, a_disp)
#plt.plot(ca, a_disp, linewidth=1.5)
##plt.plot(mutations, mutations_y, 'bx', label='Mutation')

plt.legend(loc='center', fontsize='small')

plt.axhline(y=0, color='green', linestyle='--', linewidth=1, label='y=0')

plt.title('1A53-2: Avg Displacement (2SD) with increased sampling')
plt.xlabel('Residue')
plt.ylabel('C displacements / ')
plt.ylim(-0.2,0.85)
```
